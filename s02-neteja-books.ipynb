```python
>>> import pandas as pd
>>> import numpy as np

file_name = "./dat/BL-Flickr-Images-Book.csv"
df = pd.read_csv(file_name)
df.head()
```

```python
# 1.	Eliminació de columnes en un dataframe.
to_drop = ['Edition Statement',
         'Corporate Author',
         'Corporate Contributors',
         'Former owner',
         'Engraver',
         'Contributors',
         'Issuance type',
         'Shelfmarks']

df.drop(to_drop, inplace=True, axis=1)
```

```python
### NOTA PER MI, ALS ALUMNES PASSR NOMÉS FINS AQUI
df.head()
```

```python
# 2.	Canviar l’index d’un dataframe
df = df.set_index('Identifier')
```

```python
df.head()
```

```python
#3.	Endreçar camps a les dades
df.loc[1851:, 'Date of Publication'].head(5)
```

```python
#4.	Combinant mètodes str amb numpy per netejar columnes
pub = df['Place of Publication']
london = pub.str.contains('London')
oxford = pub.str.contains('Oxford')
london[:5]
oxford[:5]
```

```python
df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))
```

```python
df.head(100)
```

```python
df = df.dropna()
```
