```python
## Excercici diapo 32 w3schools
import pandas as pd

#--------------------------------- llegeix arxiu
f = open('./dat/w3_data.csv', "r", encoding="UTF-8")
data = f.read()
f.close()

data = data.replace(",\n", "\n")
data = data.replace("'", "")
data = data.replace("NaN", "")
if data[-1] == ",":
    data = data[:-1]
#--------------------------------- Grava arxiu
f = open('./dat/w3_data_new.csv', "w", encoding="UTF-8")
f.write(data)
f.close()
```

```python
df = pd.read_csv('./dat/w3_data_new.csv')
df
```

```python
# Eliminar els espais dels noms de les columnes
# usant un diccionari amb el nom incorrecte : nom correcte
cols = df.columns.values.tolist()
d = dict()
for column in cols :
    d[column] = column.strip()
    
print(d)
df = df.rename(columns=d)
```

```python
# Quins tipus de columnes tenim ?  Compte tenim columnes de tipus erroni.
df.dtypes
```

```python
df.describe()
```

```python
# Metode 1 :Anem a arreglar la columna de Calories
df.Calories = pd.to_numeric(df.Calories, errors='coerce')
df
```

```python
# Es rremplenen els nulls amb la mitja de les calories
df.Calories.fillna(df.Calories.mean(), inplace=True)
df
```

```python
# Metode 1 :Anem a arreglar la columna de Date
df.loc[22, 'Date'] = '2020-12-22'
df.Date = pd.to_datetime(df.Date, errors='coerce')
df
```

```python
df.dtypes
# Ara s que tenim els tipus correctes
```

```python
# Valors fora de rang, posem el mxim
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120
df
```
