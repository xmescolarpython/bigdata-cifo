```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_title('Boxplot per Preu')
ax.boxplot(df["price"], labels=["Preu"], showfliers = False)

#Valors Extrems – Quartils i Outliers
q_low = df["price"].quantile(0.5)
q_hi = df["price"].quantile(0.95)
print (q_low, q_hi)
```

```python
df_filtered = df[(df["price"] > q_low) & (df["price"] < q_hi)]
df_filtered['price'].plot()
fig, ax = plt.subplots()
ax.set_title('Boxplot per Preu')
ax.boxplot(df_filtered["price"], labels=["Preu"])
```

```python
corr = df.corr ()
corr.style.background_gradient (cmap = 'coolwarm')
```
